2. Mencari Outlier

```python
# Membuat boxplot untuk setiap kolom numerik di dataframe data_survey untuk melihat distribusi data
for column in data_survey.select_dtypes(include=['number']).columns:
  plt.figure(figsize=(6, 4))
  sns.boxplot(x=data_survey[column])
  plt.title(f'Boxplot dari kolom {column}')
  plt.show()
```

3. Kategorisasi Data

> karena tidak ada outlier, maka dilakukan kategorisasi data

> Membuat variabel dengan nama psikologi_var untuk menyatukan kolom yang berpengaruh ke psikologi, seperti kolom Distraction, Restlessness, dan SleepIssues. Lalu menambahkan kolom baru yang bernilai total dari variabel psikologi_var, dengan nama kolom Psi atau psikologi index

```python

psikologi_var = ['Distraction', 'Restlessness', 'EasilyDistracted', 'Worries', 
                  'Concentration', 'Comparison', 'ComparisonFeel', 'Validation', 'Depression',
                  'InterestFluctuation', 'SleepIssues']

data_survey['Psi'] = data_survey[psikologi_var].sum(axis=1)
data_survey
```

```python
batas_rendah = 20 # membuat variabel batas_rendah dengan nilai 20
batas_tinggi = 40 # membuat variabel batas_rendah dengan nilai 40

data_survey['Psychologylevel'] = pd.cut(data_survey['Psi'], # membuat kolom dengan nama Psychologylevel dengan membagi nilai dari kolom Psi
                         bins=[10, batas_rendah, batas_tinggi, 60],  # menentukan batas nilai
                         labels=['Rendah', 'Sedang', 'Tinggi'],  # menentukan label untuk kategori
                         include_lowest=True)
data_survey 
```

```python
kolom_numerik = ['Age', 'NoPurposeUse', 'Distraction', 'Restlessness', 'EasilyDistracted',
                 'Worries', 'Concentration', 'Comparison', 'ComparisonFeel', 'Validation',
                 'Depression', 'InterestFluctuation', 'SleepIssues']

# Looping untuk membuat histogram setiap kolom numerik
for kolom in kolom_numerik:
    plt.figure(figsize=(10, 6))  # Atur ukuran figure
    sns.histplot(data=data_survey, x=kolom, kde=True)  # Buat histogram

    plt.title(f'Histogram {kolom}')
    plt.xlabel(kolom)
    plt.ylabel('Frekuensi')

    plt.show()
```

```python
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler

kolom_numerik = ['Age', 'NoPurposeUse', 'Distraction', 'Restlessness', 'EasilyDistracted',
                 'Worries', 'Concentration', 'Comparison', 'ComparisonFeel', 'Validation',
                 'Depression', 'InterestFluctuation', 'SleepIssues']


data_trans = data_survey.copy()
tr = 'log'

for i in kolom_numerik:
    if tr == 'log':
        data_trans[i] = np.log1p(data_trans[i])
    if tr == 'boxcox':
        data_trans[i],_ = stats.boxcox(data_trans[i])
    if tr == 'standar':
        data_trans[i] = StandardScaler.fit_transform(data_trans[i])
    if tr == 'minmax':
        data_trans[i] = MinMaxScaler.fit_transform(data_trans[i])
    
data_trans
```

```python
kolom = 2
baris = (len(kolom_numerik) * 2 + 1)//kolom

plt.figure(figsize=(12,4*baris))

for i, col in enumerate(kolom_numerik):
    plt.subplot(baris, kolom, 2 * i + 1)
    sns.histplot(data_survey[col], bins = 20, kde = True, label = 'Sebelum di transformasi')
    plt.title(f'distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()
    
    
    plt.subplot(baris, kolom, 2 * i + 2)
    plt.title(f'distribusi {col}')
    sns.histplot(data_trans[col], bins = 20, kde = True, label = 'Sesudah di transformasi')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()

plt.tight_layout()
plt.show()
```
